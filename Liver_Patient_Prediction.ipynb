## ***5. Hypothesis Testing***

### Hypothetical Statement - 1

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
male_totalilirubin = df[df['Gender of the patient'] == 0]['Totalilirubin']
female_totalilirubin = df[df['Gender of the patient'] == 1]['Totalilirubin']

t_stat, p_value = stats.ttest_ind(male_totalilirubin, female_totalilirubin, equal_var=False)

print(f"T-Test: {t_stat:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in Totalilirubin levels between genders.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in Totalilirubin levels between genders.")
```

### Hypothetical Statement - 2

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value

male_alkaline_phosphotase = df[df['Gender of the patient'] == 0]['Alkaline Phosphotase']
female_alkaline_phosphotase = df[df['Gender of the patient'] == 1]['Alkaline Phosphotase']

t_stat, p_value = stats.ttest_ind(male_alkaline_phosphotase, female_alkaline_phosphotase, equal_var=False) # Welch's t-test for unequal variances

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in Alkaline Phosphotase levels between genders.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in Alkaline Phosphotase levels between genders.")
```

### Hypothetical Statement - 3

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value

male_albumin = df[df['Gender of the patient'] == 0]['Albumin']
female_albumin = df[df['Gender of the patient'] == 1]['Albumin']

t_stat, p_value = stats.ttest_ind(male_albumin, female_albumin, equal_var=False) # Welch's t-test for unequal variances

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in Albumin levels between genders.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in Albumin levels between genders.")
```
